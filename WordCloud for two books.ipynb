```python
import numpy as np
import pandas as pd

import wordcloud
import requests
from bs4 import BeautifulSoup
import re
import PIL
import wptools
%matplotlib inline

import matplotlib.pyplot as plt
```

### Daniel Kahneman Think Fast Think Slow New York Times Review
Two brains running by Jim Holt

```python
dk = 'https://www.nytimes.com/2011/11/27/books/review/thinking-fast-and-slow-by-daniel-kahneman-book-review.html'
r = requests.get(dk).content
```

```python
soup = BeautifulSoup(r)
a = soup.body
text = a.find('section', class_ = 'meteredContent css-1r7ky0e')
text = str(text)
text = re.sub("<.*?>","", text)
plt.figure(figsize = [15,10])
wc = wordcloud.WordCloud(min_font_size=4, width = 800, height = 400,
                         max_font_size=50, background_color="white", colormap = 'inferno').generate(text)
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
img = np.array(PIL.Image.open("img1.png"))
```

```python
plt.figure(figsize = [15,10])
wc = wordcloud.WordCloud(min_font_size=4, width = 800, height = 400,
                         max_font_size=50, background_color="white", colormap = 'inferno',
                        mask = img).generate(text)
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

#### To the edge of the world

- Extraer revisiones de la goodreads https://www.goodreads.com/book/show/142050.This_Thing_of_Darkness
- armar dataframe con revision, y likes
- tomar las top3
- armar wordcloud con figura de barco

```python
cm = 'https://www.goodreads.com/book/show/142050.This_Thing_of_Darkness'
c = requests.get(cm).content
soup = BeautifulSoup(c)
soup.find('div', class_ = 'bookReviewsPaginationCount')
subsoup = soup.find('div', id = 'bookReviews')
a = subsoup.find_all('span', id = re.compile("^freeText"), style = 'display:none')
print(len(a))
print(type(a))
list = []
for i in a:
    i = str(i)
    list.append(i)
text = str()
for i in range(len(list)):
    text +=list[i]
text = re.sub("<.*?>","", text)
plt.figure(figsize = [15,10])
wc = wordcloud.WordCloud(min_font_size=4, width = 800, height = 400,
                         max_font_size=50, background_color="white", colormap = 'inferno').generate(text)
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
