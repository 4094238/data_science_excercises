```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import wordcloud
import requests
from bs4 import BeautifulSoup
import re
import PIL
import wptools
%matplotlib inline
```

### Daniel Kahneman Think Fast Think Slow New York Times Review
Two brains running by Jim Holt

```python
dk = 'https://www.nytimes.com/2011/11/27/books/review/thinking-fast-and-slow-by-daniel-kahneman-book-review.html'
r = requests.get(dk).content
```

```python
soup = BeautifulSoup(r)
a = soup.body
a
```

```python
text = a.find('section', class_ = 'meteredContent css-1r7ky0e')
text
```

```python
text = str(text)
text
```

```python
text = re.sub("<.*?>","", text)
```

```python
text
```

```python
plt.figure(figsize = [15,10])
wc = wordcloud.WordCloud(min_font_size=4, width = 800, height = 400,
                         max_font_size=50, background_color="white", colormap = 'inferno').generate(text)
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
img = np.array(PIL.Image.open("img1.png"))
img
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
img_tr = np.ndarray((img.shape[0], img.shape[1]), np.int32)
for i in range(len(img)):
    img_tr[i] = list(map(transform_format, img[i]))
```

```python
img_tr
```

```python
plt.figure(figsize = [15,10])
wc = wordcloud.WordCloud(min_font_size=4, width = 800, height = 400,
                         max_font_size=50, background_color="white", colormap = 'inferno',
                        mask = img).generate(text)
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
